# Featching all page data

```python
df  = pd.DataFrame()
```

```python
df
```

```python
for i in range(1,total_page+1):
  res = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}")
  temp_df = pd.DataFrame(res.json()['results'])[['id','title','overview','release_date','popularity']]
  df = pd.concat([df,temp_df],ignore_index=True)
```
